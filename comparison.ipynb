```python
from map import Map 
from dijkstra import Dijkstra
from AStar import AStar
from potentialFields import PotentialFields
```

```python
dijkstra_times = []
aStar_times = []
potentialFields_times = []

for i in range(10):
    map = Map(100, 100)
    # map.generate_random_obstacles(30)
    map.generate_random_start()
    map.generate_random_goal()

    dijkstra = Dijkstra()
    aStar = AStar()
    potentialFields = PotentialFields()

    _, dijkstraSolveTime = dijkstra.solvePath(map, map.start, map.goal)
    _, AStarSolveTime = aStar.solvePath(map, map.start, map.goal)
    _, potentialFieldSolveTime = potentialFields.solvePath(map, map.start, map.goal)

    dijkstra_times.append(dijkstraSolveTime)
    aStar_times.append(AStarSolveTime)
    potentialFields_times.append(potentialFieldSolveTime)

    print(f"Finished iteration {i+1} of 10")

print(f"Dijkstra average time: {sum(dijkstra_times)/len(dijkstra_times)} s")
print(f"A* average time: {sum(aStar_times)/len(aStar_times)} s")
print(f"Potential Fields average time: {sum(potentialFields_times)/len(potentialFields_times)} s")
```

```python
map = Map(100, 100)
map.generate_random_goal()
map.generate_random_start()
map.generate_random_obstacles(50)

dijkstra = Dijkstra()
aStar = AStar()
potentialFields = PotentialFields()

dijkstraPath, _ = dijkstra.solvePath(map, map.start, map.goal)
aStarPath, _ = aStar.solvePath(map, map.start, map.goal)
potentialFieldsPath, _ = potentialFields.solvePath(map, map.start, map.goal)

# subplots to show the different paths using map.display_map(path)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 3)
axs[0] = map.display_map(dijkstraPath, return_fig=True)
axs[0].set_title('Dijkstra')

axs[1] = map.display_map(aStarPath, return_fig=True)
axs[1].set_title('A*')

axs[2] = map.display_map(potentialFieldsPath, return_fig=True)
axs[2].set_title('Potential Fields')
```
